```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Continuous Time Signal
```

```python
t = np.linspace(0, 2, 1000)   
x = np.sin(2*np.pi*1*t)

plt.figure()
plt.plot(t, x)
plt.title("Continuous-Time Sinusoidal Signal")
plt.xlabel("Time (t)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
f = 1          # Signal frequency (Hz)
fs = 50        # Sampling frequency (Hz)

# Continuous time axis
t = np.linspace(0, 1, 1000)
x_ct = np.sin(2*np.pi*f*t)

# Discrete time index
n = np.arange(0, fs)
t_samples = n / fs
x_dt = np.sin(2*np.pi*f*t_samples)

# Plot
plt.figure()

# Continuous signal
plt.plot(t, x_ct, label="Continuous-Time Signal")

# Discrete samples
plt.stem(t_samples, x_dt, linefmt='r-', markerfmt='ro', basefmt='k', label="Sampled Signal")

plt.title("Sampling: Continuous to Discrete-Time Signal")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```

```python
# Digital Signal
# Continuous Signal → Sampled → Quantized → Digital Signal
```

```python
# Parameters
f = 5          # Signal frequency (Hz)
fs = 50        # Sampling frequency (Hz)
bits = 3       # Number of ADC bits

# Continuous signal
t = np.linspace(0, 1, 1000)
x_ct = np.sin(2*np.pi*f*t)

# Discrete sampling
n = np.arange(0, fs)
t_samples = n / fs
x_sampled = np.sin(2*np.pi*f*t_samples)

# Quantization
levels = 2**bits
x_min = -1
x_max = 1
step_size = (x_max - x_min) / levels

x_quantized = np.round((x_sampled - x_min) / step_size) * step_size + x_min

# Plot
plt.figure(figsize=(10,7))

#plt.plot(t, x_ct, label="Continuous Signal", alpha=0.6)
plt.stem(t_samples, x_sampled, linefmt='r-', markerfmt='ro', 
         basefmt='k', label="Sampled Signal")

plt.step(t_samples, x_quantized, where='mid', label="Digital Signal (Quantized)")

plt.title("Continuous → Discrete → Digital Signal")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```

```python
t = np.linspace(-2, 2, 1000)
x = np.where(t >= 0, np.exp(-t), 0)

plt.plot(t, x)
plt.title("Causal Signal")
plt.grid()
plt.show()
```

```python
t = np.linspace(-2, 2, 1000)
x = np.where(t <= 0, np.exp(t), 0)

plt.plot(t, x)
plt.title("Anti-Causal Signal")
plt.grid()
plt.show()
```

```python
t = np.linspace(-2, 2, 2000)
```

```python
x = np.where((t >= 0) & (t <= 1), 1, 0)
plt.plot(t, x)
plt.grid()
plt.show()
```

```python
x_neg = np.where((-t >= 0) & (-t <= 1), 1, 0)
plt.plot(t, x_neg)
plt.grid()
plt.show()
```

```python
x_new = (x + x_neg)
plt.plot(t, x_new)
plt.grid()
plt.show()
```

```python
x_even = (x_new) / 2
plt.plot(t, x_even)
plt.grid()
plt.show()
```

```python
x_odd = (x - x_neg) / 2
plt.plot(t, x_odd)
plt.grid()
plt.show()
```
