```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

#ODD-EVEN signal
t = np.linspace(-2, 2, 1000)
```

```python
x=np.where((t>=0) & (t<=1),1,0)
plt.plot(t,x)
plt.grid()
plt.show()
```

```python
x_neg = np.where((-t>=0) & (-t<=1),1,0)
plt.plot(t,x_neg)
plt.grid()
plt.show()
```

```python
x_even= x + x_neg
plt.plot(t,x_even)
plt.grid()
plt.show()
```

```python
#even
x_even= (x + x_neg)/2
plt.plot(t,x_even)
plt.grid()
plt.show()
```

```python
x_even= x - x_neg
plt.plot(t,x_even)
plt.grid()
plt.show()
```

```python
#odd
x_even= (x - x_neg)/2
plt.plot(t,x_even)
plt.grid()
plt.show()
```

```python
#QUANTIZATION
# ===============================
# Parameters
# ===============================
f = 5        # Signal frequency (Hz)
fs = 50      # Sampling frequency (Hz)
bits = 8     # Number of ADC bits

# ===============================
# Continuous signal
# ===============================
t = np.linspace(0, 1, 1000)
x_ct = np.sin(2 * np.pi * f * t)

# ===============================
# Discrete sampling
# ===============================
n = np.arange(0, fs)
t_samples = n / fs
x_sampled = np.sin(2 * np.pi * f * t_samples)

# ===============================
# Quantization
# ===============================
levels = 2 ** bits
x_min = -1
x_max = 1
step_size = (x_max - x_min) / levels

x_quantized = np.round((x_sampled - x_min) / step_size) * step_size + x_min

# ===============================
# Plot
# ===============================
plt.figure(figsize=(10, 7))

#plt.plot(t, x_ct, label='Continuous Signal', alpha=0.6)

plt.stem(
    t_samples,
    x_sampled,
    linefmt='r-',
    markerfmt='ro',
    basefmt='k',
    label='Sampled Signal'
)

plt.step(
    t_samples,
    x_quantized,
    where='mid',
    label='Digital Signal (Quantized)'
)

plt.title('Continuous + Discrete + Digital Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')

plt.legend()
plt.grid()

plt.show()
```

```python
#CAUSAL signal
t = np.linspace(-2, 2, 1000)
x = np.where(t >= 0, np.exp(-t), 0)
plt.plot(t,x)
plt.title("Casual_Signal")
plt.grid()
plt.show()
```

```python
# Anti-causal signal
t = np.linspace(-2, 2, 1000)
x = np.where(t < 0, np.exp(t), 0)  # non-zero for t < 0
plt.plot(t, x)
plt.title("Anti-Causal Signal")
plt.grid()
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.show()
```
